```python
import numpy as np
import scipy.spatial as sp
import cv2
import time
import math
```

```python
car1Basic = np.load("car1-stats.npy")
car1Kalman = np.load("car1-stats-kalman.npy")
car2Basic = np.load("car2-stats.npy")
car2Kalman = np.load("car2-stats-kalman.npy")

print (car1Basic.shape, car1Kalman.shape, car2Basic.shape, car2Kalman.shape)

```

```python
import matplotlib.pyplot as plt
```

```python
acc = np.array([car1Basic[i,0] for i in range(945)])
rbs = np.array([car1Basic[i,1] for i in range(945)])
x = np.asarray(range (945))
plt.rcParams['axes.facecolor'] = 'white'
plt.tick_params(axis='x', colors='black')
plt.tick_params(axis='y', colors='black')
plt.rcParams['grid.color'] = 'k'
plt.rcParams['grid.linestyle'] = ':'
plt.rcParams['grid.linewidth'] = 0.5

plt.plot(x,acc, label="Overlapping", linestyle="--", linewidth=4)
plt.plot(x,rbs, label="Robustness", linestyle="-", linewidth=4)

plt.legend()
plt.show()
```
